```python
#median developer saries by age
ages_x=[22,24,26,28,30,32,34,36,38,40,42]
dev_y=[38000,40000,42222,44444,46555,48777,50999,52111,54333,56555,58777]

#median python developers salaries by age
py_dev_y=[45444,47666,49888,51000,53222,55444,57666,59888,61555,63222,65444]

#median javascript developers salaries by age
js_dev_y=[40111,42333,44555,46777,48999,50111,52333,54555,56777,58999,60111]
```

```python
plt.plot(ages_x,dev_y,label="All devs")
plt.plot(ages_x,py_dev_y,label="Python")
plt.xlabel("Ages")
plt.ylabel("Median Salary(USD)")
plt.title("Median Salary(USD)by age")
plt.legend()
```

```python
#stacked
plt.bar(ages_x,py_dev_y,color="#008df5",label="Python")
plt.bar(ages_x,dev_y,color="#555555",label="All devs")
plt.bar(ages_x,js_dev_y,color="#e5ae38",label="Javascript")
plt.xlabel("Ages")
plt.ylabel("Median Salary(USD)")
plt.title("Median Salary(USD) by age")
plt.legend()
#Automatically adjust subplot parameters to give specified padding
plt.tight_layout()
plt.show()
```

```python
#side by side
x_indexes=np.arange(len(ages_x))
bar_width=0.25
plt.bar(x_indexes-bar_width,dev_y,width=bar_width,color="#444444",label="All devs")
plt.bar(x_indexes,py_dev_y,width=bar_width,color="#008df5",label="Python")
plt.bar(x_indexes+bar_width,js_dev_y,width=bar_width,color="#e5ae38",label="Javascript")
        
#without this line our age range is lost,to fix this:
plt.xticks(ticks=x_indexes,labels=ages_x)
plt.xlabel("Ages")
plt.ylabel("Median salary (USD)")
plt.title("Median salary(USD) by Age")
plt.legend()
plt.tight_layout()
plt.show()
```
