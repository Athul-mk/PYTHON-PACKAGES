```python
#graphical representation of sin
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show
```

```python
# graphical represention of cos
x=np.arange(0,5*np.pi,0.5)
y=np.cos(x)
plt.plot(x,y)
plt.show
```

```python
#graphical representation of tan
x=np.arange(0,10*np.pi,0.1)
y=np.tan(x)
plt.plot(x,y)
plt.show
```
